```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from joblib import dump

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    roc_auc_score,
    classification_report,
    mean_absolute_error,
    confusion_matrix,
    ConfusionMatrixDisplay
)

from sklearn.preprocessing import MinMaxScaler  # <-- Ensure we import MinMaxScaler
from imblearn.over_sampling import SMOTE
from catboost import CatBoostClassifier

# -----------------------------------------------------------------------------
# 0. PREPARE DATA
# -----------------------------------------------------------------------------
# Suppose `df` is your DataFrame containing all feature subsets + 'bad_flag'.
df = df.fillna(0)

# Ensure target variable exists
if 'bad_flag' not in df.columns:
    raise ValueError("The target column 'bad_flag' is not in the dataset.")

# Define your column groups
transaction_cols = [col for col in df.columns if col.startswith("transaction_attribute")]
bureau_cols = [col for col in df.columns if col.startswith("bureau")]
bureau_enquiry_cols = [col for col in df.columns if col.startswith("bureau_enquiry")]
onus_cols = [col for col in df.columns if col.startswith("onus")]

# -----------------------------------------------------------------------------
# 1. SINGLE TRAIN/TEST SPLIT FOR THE WHOLE DATASET
# -----------------------------------------------------------------------------
X = df.drop(columns=['bad_flag'])
y = df['bad_flag']

X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale the data AFTER splitting (to avoid data leakage)
scaler = MinMaxScaler()
X_train_full_scaled = scaler.fit_transform(X_train_full)
X_test_full_scaled = scaler.transform(X_test_full)

# Convert scaled arrays back to DataFrames (so we can select columns by name)
X_train_full_scaled_df = pd.DataFrame(
    X_train_full_scaled,
    columns=X_train_full.columns,
    index=X_train_full.index
)
X_test_full_scaled_df = pd.DataFrame(
    X_test_full_scaled,
    columns=X_test_full.columns,
    index=X_test_full.index
)

# Calculate scale_pos_weight once
scale_pos_weight = len(y_train_full[y_train_full == 0]) / len(y_train_full[y_train_full == 1])

# Common CatBoost params
best_params_cat = {
    'iterations': 100,
    'depth': 4,
    'learning_rate': 0.1,
    'random_seed': 42,
    'l2_leaf_reg': 3,
    'thread_count': -1,
    'eval_metric': 'AUC'
}

# -----------------------------------------------------------------------------
# 2. FUNCTION TO TRAIN A SINGLE MODEL GIVEN COLUMN SUBSET
# -----------------------------------------------------------------------------
def train_catboost_with_smote(X_train_subset, X_test_subset, y_train, y_test):
    """
    Given training/test subsets of features (X) and target (y),
    this function applies SMOTE, trains CatBoost, and returns:
      - The trained model
      - The predictions (probabilities) on the test set
    """
    # SMOTE Resampling (uncomment if you want to actually oversample):
    # smote = SMOTE(random_state=42)
    # X_resampled, y_resampled = smote.fit_resample(X_train_subset, y_train)
    #
    # model = CatBoostClassifier(
    #     **best_params_cat,
    #     class_weights=[1, scale_pos_weight]
    # )
    # model.fit(
    #     X_resampled, y_resampled,
    #     eval_set=(X_test_subset, y_test),
    #     use_best_model=True,
    #     verbose=False
    # )

    # If not using SMOTE, just train directly:
    model = CatBoostClassifier(
        **best_params_cat,
        class_weights=[1, scale_pos_weight]
    )
    model.fit(
        X_train_subset, y_train,
        eval_set=(X_test_subset, y_test),
        use_best_model=True,
        verbose=False
    )

    # Predict probabilities on the test set
    y_pred_prob = model.predict_proba(X_test_subset)[:, 1]
    return model, y_pred_prob

# -----------------------------------------------------------------------------
# 3. TRAIN FOUR MODELS ON THEIR RESPECTIVE FEATURE SUBSETS (ON SCALED DATA)
# -----------------------------------------------------------------------------
# 3.1 Transaction Attributes
X_train_trn = X_train_full_scaled_df[transaction_cols]
X_test_trn = X_test_full_scaled_df[transaction_cols]
model_transaction, y_pred_transaction = train_catboost_with_smote(
    X_train_trn, X_test_trn, y_train_full, y_test_full
)
auc_trn = roc_auc_score(y_test_full, y_pred_transaction)
mae_trn = mean_absolute_error(y_test_full, y_pred_transaction) * 100
print(f"[Transaction] AUC: {auc_trn:.4f}, MAE: {mae_trn:.4f}")

# 3.2 Bureau Attributes
X_train_bur = X_train_full_scaled_df[bureau_cols]
X_test_bur = X_test_full_scaled_df[bureau_cols]
model_bureau, y_pred_bureau = train_catboost_with_smote(
    X_train_bur, X_test_bur, y_train_full, y_test_full
)
auc_bur = roc_auc_score(y_test_full, y_pred_bureau)
mae_bur = mean_absolute_error(y_test_full, y_pred_bureau) * 100
print(f"[Bureau] AUC: {auc_bur:.4f}, MAE: {mae_bur:.4f}")

# 3.3 Bureau Enquiry Attributes
X_train_bur_enq = X_train_full_scaled_df[bureau_enquiry_cols]
X_test_bur_enq = X_test_full_scaled_df[bureau_enquiry_cols]
model_bureau_enq, y_pred_bureau_enq = train_catboost_with_smote(
    X_train_bur_enq, X_test_bur_enq, y_train_full, y_test_full
)
auc_bur_enq = roc_auc_score(y_test_full, y_pred_bureau_enq)
mae_bur_enq = mean_absolute_error(y_test_full, y_pred_bureau_enq) * 100
print(f"[Bureau Enquiry] AUC: {auc_bur_enq:.4f}, MAE: {mae_bur_enq:.4f}")

# 3.4 Onus Attributes
X_train_onus = X_train_full_scaled_df[onus_cols]
X_test_onus = X_test_full_scaled_df[onus_cols]
model_onus, y_pred_onus = train_catboost_with_smote(
    X_train_onus, X_test_onus, y_train_full, y_test_full
)
auc_onus = roc_auc_score(y_test_full, y_pred_onus)
mae_onus = mean_absolute_error(y_test_full, y_pred_onus) * 100
print(f"[Onus] AUC: {auc_onus:.4f}, MAE: {mae_onus:.4f}")

# -----------------------------------------------------------------------------
# 4. PRINT INDIVIDUAL CONFUSION MATRICES
# -----------------------------------------------------------------------------
y_pred_transaction_bin = (y_pred_transaction > 0.5).astype(int)
cm_trn = confusion_matrix(y_test_full, y_pred_transaction_bin)
disp_trn = ConfusionMatrixDisplay(confusion_matrix=cm_trn, display_labels=["No Default", "Default"])
disp_trn.plot(cmap="Blues")
plt.title("Confusion Matrix: Transaction Attributes")
plt.show()

y_pred_bureau_bin = (y_pred_bureau > 0.5).astype(int)
cm_bur = confusion_matrix(y_test_full, y_pred_bureau_bin)
disp_bur = ConfusionMatrixDisplay(confusion_matrix=cm_bur, display_labels=["No Default", "Default"])
disp_bur.plot(cmap="Blues")
plt.title("Confusion Matrix: Bureau Attributes")
plt.show()

y_pred_bureau_enq_bin = (y_pred_bureau_enq > 0.5).astype(int)
cm_bur_enq = confusion_matrix(y_test_full, y_pred_bureau_enq_bin)
disp_bur_enq = ConfusionMatrixDisplay(confusion_matrix=cm_bur_enq, display_labels=["No Default", "Default"])
disp_bur_enq.plot(cmap="Blues")
plt.title("Confusion Matrix: Bureau Enquiry Attributes")
plt.show()

y_pred_onus_bin = (y_pred_onus > 0.5).astype(int)
cm_onus = confusion_matrix(y_test_full, y_pred_onus_bin)
disp_onus = ConfusionMatrixDisplay(confusion_matrix=cm_onus, display_labels=["No Default", "Default"])
disp_onus.plot(cmap="Blues")
plt.title("Confusion Matrix: Onus Attributes")
plt.show()

# -----------------------------------------------------------------------------
# 5. ENSEMBLE STEP: AVERAGE THE PREDICTIONS FROM ALL MODELS
# -----------------------------------------------------------------------------
y_pred_ensemble_prob = (
    y_pred_transaction +
    y_pred_bureau +
    y_pred_bureau_enq +
    y_pred_onus
) / 4.0

y_pred_ensemble_bin = (y_pred_ensemble_prob > 0.5).astype(int)

# -----------------------------------------------------------------------------
# 6. EVALUATE THE ENSEMBLED MODEL
# -----------------------------------------------------------------------------
auc_score_ensemble = roc_auc_score(y_test_full, y_pred_ensemble_prob)
mae_ensemble = mean_absolute_error(y_test_full, y_pred_ensemble_prob) * 100

print(f"\n=== ENSEMBLE RESULTS ===")
print(f"Ensemble AUC: {auc_score_ensemble:.4f}")
print(f"Ensemble MAE: {mae_ensemble:.4f}\n")

print("Classification Report (ENSEMBLE):")
print(classification_report(y_test_full, y_pred_ensemble_bin))

cm_ensemble = confusion_matrix(y_test_full, y_pred_ensemble_bin)
disp_ensemble = ConfusionMatrixDisplay(confusion_matrix=cm_ensemble, display_labels=["No Default", "Default"])
disp_ensemble.plot(cmap="Blues")
plt.title("Confusion Matrix (Ensemble)")
plt.show()
import os
os.makedirs("../models", exist_ok=True)
dump(model_transaction, "../models/model_transaction.joblib")
dump(model_bureau, "../models/model_bureau.joblib")
dump(model_bureau_enq, "../models/model_bureau_enq.joblib")
dump(model_onus, "../models/model_onus.joblib")
```
